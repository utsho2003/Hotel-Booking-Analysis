# **Project Summary -**

This project focuses on analyzing hotel booking data to understand customer behavior, booking patterns, and the factors that influence booking cancellations and hotel revenue. The main objective of this project is to perform Exploratory Data Analysis (EDA) on the hotel booking dataset and extract meaningful insights that can help hotels make better business decisions, improve customer satisfaction, and increase overall profitability.

The dataset contains detailed information about hotel bookings, including hotel type, booking status, lead time, arrival date, number of adults, children, babies, country of origin, market segment, distribution channel, average daily rate (ADR), customer type, and reservation status. The dataset includes both categorical and numerical variables, which provide valuable information about customer demographics, booking trends, and pricing strategies.

The first step of the project involved understanding the structure of the dataset and performing data cleaning. After cleaning the data, exploratory data analysis was performed using various visualization techniques such as count plots, bar plots, histograms, box plots, pie charts, scatter plots, correlation heatmaps, and pair plots. These visualizations helped in understanding the distribution of variables and identifying relationships between different features. For example, it was observed that City Hotels have more bookings compared to Resort Hotels. Online travel agencies were found to be the primary source of bookings, highlighting the importance of online platforms in the hotel industry.

The analysis also revealed that a significant number of bookings were canceled, which can negatively impact hotel revenue. It was found that bookings with longer lead times had a higher probability of cancellation. Additionally, most customers were identified as transient customers, meaning they were not repeat guests. This indicates an opportunity for hotels to improve customer loyalty programs and retention strategies.

The correlation heatmap helped identify relationships between numerical variables such as lead time, ADR, and number of adults. The pair plot provided a deeper understanding of how these variables interact with each other and how they influence booking outcomes. These insights can help hotels optimize pricing strategies, improve resource management, and reduce cancellation rates.

Overall, this project provided valuable insights into hotel booking trends, customer behavior, and business performance. The findings can help hotel management make data-driven decisions, improve operational efficiency, enhance customer satisfaction, and increase revenue. By leveraging these insights, hotels can develop better marketing strategies, improve booking policies, and strengthen customer relationships, ultimately contributing to long-term business growth and success.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df.shape
df.isnull().sum()

# Fill children missing values with 0
df['children'].fillna(0, inplace=True)

# Fill agent missing values with 0
df['agent'].fillna(0, inplace=True)

# Fill company missing values with 0
df['company'].fillna(0, inplace=True)

# Fill country missing values with mode
df['country'].fillna(df['country'].mode()[0], inplace=True)

# 4. Remove duplicate rows
df.drop_duplicates(inplace=True)

# 5. Convert data types if needed
df['children'] = df['children'].astype(int)

df.isnull().sum()
````

### What all manipulations have you done and insights you found?

Manipulations done:
Handled missing values, removed duplicate records, corrected data types, and explored unique values to clean and prepare the dataset for analysis.

Insights found:
City Hotels had more bookings, many bookings were canceled, most customers were new guests, and online travel agencies were the main booking source.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

plt.figure(figsize=(12,8))

# select only numerical columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()

````

##### 1. Why did you pick the specific chart?
